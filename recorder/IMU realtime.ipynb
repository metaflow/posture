```python
last_direction = np.zeros(3)
ser = None
def direction():
    global ser
    if not ser:
        ser = serial.Serial(serial_port, 38400)
    global last_direction
    ln = ser.readline().decode('utf-8')
#     print(ln)
    parts = ln.strip().split('\t')
    while ser.inWaiting() > 0:
        ln = ser.readline().decode('utf-8')
#         print(ln)
        p = ln.strip().split('\t')
        if len(p) >= 5:
            parts = p
    if len(parts) >= 5:
        last_direction[0] = parts[3]
        last_direction[1] = parts[4]
        last_direction[2] = parts[5]
    return last_direction
```
